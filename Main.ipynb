````python
#Making a class for Binary Image Classification
class BinaryClassCNN(nn.Module): #CNN is for Convolutional Neural Network
    def __init__(self):
        super(BinaryClassCNN, self).__init__()
        self.conv1 = nn.Conv2d(1, 16, kernel_size= 3, stride= 1, padding=1) #How the pixels are set up in the convolutional layer
        self.relu = nn.ReLU()
        self.pool = nn.MaxPool2d(kernel_size=2, stride = 2) #How the identifier? is made and moves over the convolutional layer
        self.flatten = nn.Flatten()
        self.fc1 = nn.Linear(16, 122, 122, 1) # This is the Linear layer, can be customized, the 1 shows the output of self pool I believe
        self.sigmoid = nn.Sigmoid() # activation function
        
    def forward(self, x):   #This combines all of BinaryClassCNN -> forward pass
        x = self.pool(self.relu(self.conv1(x)))
        x = self.fc1(self.flatten(x))
        x = self.sigmoid(x)
        return x
````

````python
#Adding convolutional layers to the model

class Conv(nn.Module):
    def __init__(self):
        super(Conv, self).__init__()
        self.conv1 = nn.Conv2d(in_channels=3, out_channels=16, kernel_size=3, padding=1)
        
conv2 = nn.Conv2d(in_channels=16, out_channels=32, kernel_size=3, padding=1)

model = Conv()
model.add_module('conv2', conv2)
````

````python
#Stacking convolutional layers in a block

class BinaryImageClassification(nn.Module):
    def __init__(self):
        super(BinaryImageClassification, self).__init__()
        self.conv_block = nn.Sequential (#Activation feature
        nn.Conv2d(3,16, kernel_size=3, stride=1, padding=1), #Values from the Conv Class
        nn.Relu(), #another activiation feature
        nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=1),
        nn.ReLu(),
        nn.MaxPoll2d(kernel_size=2, stride=2)
        )
    def forward(self, x):
        x = self.conv_block(x)
````
